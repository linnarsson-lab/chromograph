```python
import os, sys, glob

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import loompy
import logging
import glob

logger = logging.getLogger()
logging.basicConfig(
    format='%(asctime)s %(levelname)-8s %(message)s',
    level=logging.INFO,
    datefmt='%H:%M:%S')
```

```python
from chromograph.pipeline.utils import *
from cytograph.pipeline.config import load_config
RNA_config = load_config()
from chromograph.pipeline.config import load_config
ATAC_config = load_config()
ATAC_config.params.max_fragments = 1e6
ATAC_config.params.FR_TSS = 0.1
ATAC_config.params.level = 1000
ATAC_config.paths.samples = '/data/proj/AllenInst/ATAC/'

samples = ['Allen_H19_30_001', 'Allen_H19_30_002']

print(ATAC_config.params)

ATAC_dir = ATAC_config.paths.samples
RNA_dir = '/data/proj/AllenInst/RNA'

f_ATAC = '/data/proj/AllenInst/matched_analysis/ATAC/All/All.loom'
f_RNA = '/data/proj/AllenInst/matched_analysis/RNA/data/All.loom'

ATAC_selections = []
RNA_selections = []

for s in samples:
    with loompy.connect(os.path.join(ATAC_dir, f'10X{s}', f'10X{s}_20kb.loom')) as ds:
        good_cells = ((ds.ca.DoubletFinderFlag == 0) & 
                      (ds.ca.passed_filters > ATAC_config.params.level) & 
                      (ds.ca.passed_filters < ATAC_config.params.max_fragments) & 
                      (ds.ca.TSS_fragments/ds.ca.passed_filters > ATAC_config.params.FR_TSS))
        ATAC_CellIDs = ds.ca.CellID[good_cells]
        
    with loompy.connect(os.path.join(RNA_dir, f'10X{s}.loom')) as ds:
        predicted_doublets = ds.ca.DoubletFinderFlag
        high_UMI = ds.ca.TotalUMI >= RNA_config.params.min_umis
        low_MT = ds.ca.MT_ratio < RNA_config.params.max_fraction_MT_genes
        high_unspliced = ds.ca.unspliced_ratio > RNA_config.params.min_fraction_unspliced_reads
        good_cells = np.all([high_UMI, low_MT, high_unspliced], axis = 0)
        
        RNA_CellIDs = ds.ca.CellID[good_cells]
        
    valid_cells = ATAC_CellIDs[np.isin(ATAC_CellIDs,RNA_CellIDs)]
    
    with loompy.connect(os.path.join(ATAC_dir, f'10X{s}', f'10X{s}_20kb.loom')) as ds:
        ATAC_selections.append(np.array([x in valid_cells for x in ds.ca.CellID]))
    with loompy.connect(os.path.join(RNA_dir, f'10X{s}.loom')) as ds:
        RNA_selections.append(np.array([x in valid_cells for x in ds.ca.CellID]))
        
logging.info(f'Generate ATAC file')
skip_attr = find_attr_to_skip(ATAC_config, samples)
skip_attr = set(ATAC_config.params.skip_attrs + skip_attr)
inputfiles = [os.path.join(ATAC_dir, f'10X{s}', f'10X{s}_20kb.loom') for s in samples]
loompy.combine_faster(inputfiles, f_ATAC, selections=ATAC_selections, key = 'loc', skip_attrs=skip_attr)

logging.info(f'Generate RNA file')
inputfiles = [os.path.join(RNA_dir, f'10X{s}.loom') for s in samples]
loompy.combine_faster(inputfiles, f_RNA, {}, RNA_selections, key="Accession", skip_attrs=["_X", "_Y", "Clusters"])
with loompy.connect(f_RNA) as ds:
    ds.ca.SampleID = np.array([x.split(':')[0] for x in ds.ca.CellID])
```

```python
def rna_barcodes_to_atac(ds):
    barcodes_rna = []
    for barcode in ds.ca.CellID:
        sample = barcode.split(':')[0]
        if not sample[:3] == '10X':
            sample = '10X' + sample
        b = barcode.split(':')[-1][:-1]
        barcodes_rna.append(':'.join([sample,b]))
    return barcodes_rna

samples = ['10XAllen_H19_30_001', '10XAllen_H19_30_002']

for sample in samples:
    RNA = f'/data/proj/AllenInst/RNA/{sample}.loom'
    bins = f'/data/proj/AllenInst/ATAC/{sample}/{sample}_5kb.loom'
    
    with loompy.connect(bins) as ds:
        with loompy.connect(RNA) as dsr:
            if not 'DoubletFinderScore' in dsr.ca:
                logging.info(f'ERROR: NO DOUBLETFINDERSCORE IN RNA!')
            RNA_bar = rna_barcodes_to_atac(dsr)
            print(np.sum(np.isin(RNA_bar, ds.ca.CellID)))
            print(RNA_bar[:5])
            
            match = {k:v for v, k in enumerate(ds.ca.CellID)}

            ## Initialize empty attributes
            dbs, dbf = np.zeros((ds.shape[1],)).astype(np.float64), np.zeros((ds.shape[1],)).astype(np.int64)
            tsne = np.zeros((ds.shape[1],2)).astype(np.float64)

            for i, x in enumerate(RNA_bar):
                k = match[x]
                dbs[k] = dsr.ca['DoubletFinderScore'][i]
                dbf[k] = dsr.ca['DoubletFinderFlag'][i]
                tsne[k] = dsr.ca['TSNE'][i]

            ds.ca['DoubletFinderScore'] = dbs
            ds.ca['DoubletFinderFlag'] = dbf
            ds.ca['TSNE'] = tsne
```

```python
from cytograph.manifold import BalancedKNN
from scipy import sparse

k = 25

with loompy.connect('/data/proj/AllenInst/ATAC_builds/ATAC_build_Harmony/subset/subset_peaks.loom') as ds:
    print(ds.ca.keys())
    decomp = ds.ca.PCA
    metric = 'euclidean'
    
    ## Construct nearest-neighbor graph
    logging.info(f"Computing balanced KNN (k = {k}) space using the '{metric}' metric")
    bnn = BalancedKNN(k=k, metric=metric, maxl=2 * k, sight_k=2 * k, n_jobs=-1)
    bnn.fit(decomp)
    knn = bnn.kneighbors_graph(mode='distance')
    knn.eliminate_zeros()
    mknn = knn.minimum(knn.transpose())
    # Convert distances to similarities
    logging.info(f'Converting distances to similarities knn shape.')
    max_d = knn.data.max()
    knn.data = (max_d - knn.data) / max_d
    mknn.data = (max_d - mknn.data) / max_d
    ds.col_graphs.KNN = knn
    ds.col_graphs.MKNN = mknn
    mknn = mknn.tocoo()
    mknn.setdiag(0)
    # Compute the effective resolution
    logging.info(f'Computing resolution')
    d = 1 - knn.data
    radius = np.percentile(d, 90)
    logging.info(f"  90th percentile radius: {radius:.02}")
    ds.attrs.radius = radius
    inside = mknn.data > 1 - radius
    rnn = sparse.coo_matrix((mknn.data[inside], (mknn.row[inside], mknn.col[inside])), shape=mknn.shape)
    ds.col_graphs.RNN = rnn
```

```python
## Plot motifs

N = 5
layer = '-log_pval_trinaries'

lines = [3,13,26,30,37,45,58]

xlabels = ['Opc', 'Splatter', 'Exc', 'Endo/Per/VLMC', 'Astro', 'Immune', 'Oligo']
xticks = [lines[0]/2]
for i in range(1,len(lines)):
    xticks.append((lines[i]+lines[i-1]+1)/2)

with loompy.connect('/data/proj/AllenInst/ATAC_builds/ATAC_build_Harmony/All/All_motifs.agg.loom') as ds:
    if N:
        valids = []
        for i in range(ds.shape[1]):
            q = np.quantile(ds[layer][:,i], (ds.shape[0]-N)/ds.shape[0])
            valids.append(np.where(ds[layer][:,i]>q)[0])
        valids = np.unique([i for s in valids for i in s])
    else:
        valids = np.repeat(True,ds.shape[0])
    ds.ra.valids = np.zeros(ds.shape[0])
    ds.ra.valids[valids] = True
    
    mask = np.zeros(ds.shape[0], dtype=bool)
    mask[valids] = 1
    
    TF_order = np.zeros(ds.shape[0], dtype='int')
    TF_order[mask] = np.argmax(ds.layer[layer][np.where(mask)[0], :], axis=1)
    TF_order[~mask] = np.argmax(ds.layer[layer][np.where(~mask)[0], :], axis=1) + ds.shape[1]
    TF_order = np.argsort(TF_order)
    ds.permute(TF_order, axis=0)
    
    shape_factor = len(valids)/ds.shape[1]
    
    fig, ax = plt.subplots(figsize=(32,int(shape_factor*32)))
    x = np.where(ds.ra.valids)[0]
    im = ax.imshow(ds[layer][x,:], cmap = 'Reds', vmin=2, vmax=300)
    
    labels = [f'{ds.ra.TF[i]}' for i in x]
    
    for line in lines:
        ax.axvline(line+.5, c='black', linewidth=.5)
    
    ax.set_xticks(xticks)
    ax.tick_params(axis='x', rotation=45)
    ax.set_xticklabels(xlabels, fontsize=18)
    ax.set_yticks(range(len(x)))
    ax.set_yticklabels(labels, fontsize=10)
    cbar = fig.colorbar(im, ax=ax, orientation='vertical', shrink=.25)
    cbar.ax.tick_params(labelsize=18) 
    cbar.set_label('-log10 p-val', fontsize=18)
    
    plt.savefig('/data/proj/AllenInst/ATAC_builds/ATAC_build_Harmony/All/exported/TF_heatmap.png')
```
