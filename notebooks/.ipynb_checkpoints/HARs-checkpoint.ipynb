```python
## Miscelaneous
import numpy as np
import os
import sys
import collections
import matplotlib.pyplot as plt
import loompy
from collections import Counter
from tqdm import tqdm
import pickle as pkl

import pybedtools
from pybedtools import BedTool
```

```python
f_peaks = '/proj/DEV_ATAC/Build_20210610/All/peaks/Compounded_peaks.bed'
f_HARs = '/datb/sl/camiel/HARs/Tbl_S1_HARs_table_hg38.bed'
```

```python
peaks = BedTool(f_peaks)
HARs = BedTool(f_HARs)
print(f'Totals HARs: {HARs.count()}, directly overlapping peaks: {HARs.intersect(peaks).count()}')
```

```python
distances = np.array([int(x[-1]) for x in HARs.closest(peaks, d=True)])
plt.hist(np.log10(distances+1))
```

```python
f_agg = '/proj/DEV_ATAC/Build_20210610/All/All_peaks.agg.loom'

with loompy.connect(f_agg) as dsagg:
    print(dsagg.ra.keys())
    peaks = BedTool([[dsagg.ra.Chr[i], dsagg.ra.Start[i], dsagg.ra.End[i]] for i in range(dsagg.shape[0])])
    HARs = peaks.intersect(BedTool(f_HARs), wa=True)
    HARs = np.array([f'{x[0]}:{int(x[1])-1}-{x[2]}' for x in HARs])
    valids = np.array([x in HARs for x in dsagg.ra.ID])
    print(sum(valids))
```

```python
with loompy.connect(f_agg) as dsagg:
    max_residuals = []
    clusters = []
    
    for x in np.where(valids)[0]:
        max_residuals.append(np.max(dsagg['residuals'][x,:]))
        clusters.append(np.where(dsagg['residuals'][x,:] == np.max(dsagg['residuals'][x,:]))[0])
        
    clusters = [x for s in clusters for x in s]
    plt.hist(max_residuals, range=(-10,50))
```

```python
counts = Counter(clusters)
plt.scatter(counts.keys(), counts.values())
```
